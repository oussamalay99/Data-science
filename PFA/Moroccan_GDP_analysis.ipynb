### ARIMA model

#### Introduction 

One of the powerful times series  is the ARIMA class of models which based on describing autocorrelations in the data.
ARIMA stands for Autoregssive Integrated Moving Average. 

ARIMA components : 
* `d` : Order of differencing required to make the series stationary  
* `p` : Number of autoregressive 
* `q` : Number of moving average lags 
    

**N.B:**
the seasonal version of this models family contains additional parameters (SARIMA(p,d,q)(P,D,Q)) that has the same role as the other parameters but just for the seasonal components.

Now, we need to create test and train datasets in order to build ARIMA model

```python
# We are going to split our data into two parts (train - test)
train_size = int(len(df) * .66)
train, test = df["PIB (MDH)"].iloc[:train_size], df["PIB (MDH)"].iloc[train_size:]
```

```python
train.shape, test.shape
```

```python
plt.plot(train, label="train")
plt.plot(test, label= "test")
plt.legend();
```

#### Stationarity of our TS

Before moving to build the model we need to check the stationarity of our series 

As we mentioned before one of the requirements for ARIMA is time series should be stationary series  i.e  the properties do not change over time <br>there are several methods to check the stationarity<br>we will use  Augmented Dickey-Fuller test 

 Augmented Dickey-fuller test is a unit root test using autogressive model:
 * **p-value <= 0.05** : ( significance level ) null Hypothesis H0 is rejected and series is stationary 
 * **p-value >0.05** : H0 is not rejected so series isn't stationary ( requires diffrencing)

```python
# testing the stationarity of TS
print("p-value: ",sts.adfuller(train)[1])
```

The p-value is greater than 0.05 so we can't reject the H0. We are going to use the `diff` method to try to make our time series stationary.

```python
diff_1 = train.diff()
```

```python
# testing the stationarity of the transformed TS
print("p-value: ",sts.adfuller(diff_1.dropna())[1])
```

So, differencing our TS assures the stationarity. So integrating our TS once is necessary to achieve stationarity (d=1).

#### Model Parameters' analysis 

Now that we determined the d parameter of our model, we should find a way that would help us to get an idea of the other two parameters. The **ACF** and **PACF** plots will help us to achieve that.

```python
# ACF plot for our TS
sgt.plot_acf(train, lags= 17, zero=False)
plt.title("PIB (MDH)'s ACF")
plt.xlabel("Lags");
```

```python
# PACF plot For our TS
sgt.plot_pacf(train, lags= 17, zero=False)
plt.title("PIB (MDH)'s PACF")
plt.xlabel("Lags");
```

These two plots show us that using more than 5 lags to build our model is not significant.
To find the optimal model for our TS, We will use the `pmdarima.auto_arima()` function that will try several models 
and returns the one with the mininmum AIC values.
We are goin to use the results of our previous analysis to set the parameters of this function (d=1, max_p=5, max_q= 5).

#### Modeling

We can now build our model using the results of our previous analysis

```python
def arima_model(time_series):
    auto =auto_arima(time_series,
                     start_p = 0,
                     start_q = 0,
                     d =1,
                     max_p = 5,
                     max_q = 5,
                     seasonal= True,# we want to include the seasonal components in our model
                     error_action= 'warn',
                     random_state=20,
                     test ="adf",
                     D =1,
                     m =8, #after a process of try and fail we found the best value for m that is 8
                     stepwise=True,
                     suppress_warnings=True,
                     n_fits= 10,
                     trace = True)
    return auto
```

```python
arima_m = arima_model(train)
arima_m.summary()
```

So, the best model that fits our data is SARIMA(0,1,1)(1,1,0)[8]

#### Predictions

Let's use this model to make some prediction on the testing data:

```python
#Storing the predictions of our model in a DF
preds_arima =pd.DataFrame(arima_m.predict(len(test)), index = test.index)
preds_arima.columns = ['predict_PIB']
preds_arima
```

#### Model Evaluation 

Now, let's evaluate our model. From now on, we will use the `forecast_accuracy()` function that takes in parameters the predicted and the real values and provides a list of the essentiel metrics that we will use to evaluate our models. The metrics that we will use for this purpose are:
* MAPE : Mean Absolute Percentage Error
* ME  : Mean Error
* MAE  : Mean Absolute Error
* MPE : Mean Percentage Error
* RMSE : Root Mean Squared Error
* R-squared

```python
def forecast_accuracy(forecast, actual):
    mape = np.mean(np.abs(forecast - actual)/np.abs(actual))  # MAPE
    me = np.mean(forecast - actual)             # ME
    mae = np.mean(np.abs(forecast - actual))    # MAE
    mpe = np.mean((forecast - actual)/actual)   # MPE
    rmse = np.mean((forecast - actual)**2)**.5  # RMSE
    r2 = r2_score(forecast, actual)
    
    return [mape, me, mae, mpe, rmse, r2]
```

```python
scores_arima = forecast_accuracy(preds_arima["predict_PIB"], test)
score_names = ["MAPE","ME","MAE","MPE","RMSE", "R2-score"]
df_eval = pd.DataFrame([scores_arima], index=["ARIMA"], columns= score_names)
df_eval
```

#### Cross validation

now, let's try to apply the cross validation for this model:

```python
# Set the number of folds that we will use and inititate our CV splitter
n_splits = 5
tscv = TimeSeriesSplit(n_splits = n_splits)

# Initiate the variable that will store the scores
scores_cv = []

# loop over the folds: fit the model, make prdictions, store the scores
for train_index, test_index in tscv.split(df["PIB (MDH)"]):
#     Split the fold into training and testing parts
    cv_train, cv_test = df["PIB (MDH)"].iloc[train_index], df["PIB (MDH)"].iloc[test_index]
    
#     fit the model
    arima_cv = SARIMAX(cv_train, order=(0,1,1), seasonal_order=(1,1,0,8)).fit(disp=False)
    
#     predictions + scores
    predictions = arima_cv.predict(cv_test.index.values[0], cv_test.index.values[-1])
    true_values = cv_test.values
    scores_cv.append(forecast_accuracy(predictions, true_values))

#     display the results
scores_df = pd.DataFrame(scores_cv, columns= score_names, index=[f"Fold {i}" for i in range(1,n_splits+1)])
scores_cv = pd.DataFrame(np.mean(scores_cv, axis=0).reshape(1,len(score_names)), columns= score_names, index=["CV avg"])
scores_df.append(scores_cv)
```

### LSTM Model

#### Introduction

Long short-term memory (LSTM) is an artificial recurrent neural network (RNN) architecture used in the field of deep learning. Unlike standard feedforward neural networks, LSTM has feedback connections. It can not only process single data points (e.g. images), but also entire sequences of data (such as speech or video inputs).
LSTM models are able to store information over a period of time.

We are going to use this model in order to predict the value of the GDP using its past values that are contained in a sequence.

```python
#importing necessary libraries
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
```

```python
# quick view on our data
df.head()
```

#### Preprocessing 

First, we are going to split our data in two portions. One for training our model and the other for testing it.

```python
#Split data
train_lstm, test_lstm = df[:train_size][["PIB (MDH)"]], df[train_size:][["PIB (MDH)"]]
train_lstm.shape, test_lstm.shape
```

In order to boost the performance of our model, it is good to normalize our data before fitting our model. To do so, we are going to use The `MinMaxScaler` that will complete that for us.

```python
#preprocessing our data
sc = MinMaxScaler(feature_range = (0, 1))
train_lstm_scaled = sc.fit_transform(train_lstm)
```

Now, we are going to define a preprocessing specfic to the LSTM model as function that will return two arrays. The first one will be a 3-dimensional array that follows the shape (samples, time_steps, n_features), and the second one returns the target features. the first one will contain the sequences that we ll need to predict the value of each row in the target variable.  

```python
def create_dataset(X, y, time_steps=1):
    Xs, ys = [], []
    for i in range(len(X) - time_steps):
        Xs.append(X[i:i+time_steps, 0])
        ys.append(y[i+time_steps, 0])
    return np.array(Xs), np.array(ys)
```

Let's create now our training dataset

```python
# We specify the number of time steps that we ll include in each sequence
TIME_STEPS = 1

# We prepare our training and testing sets separately using the previous function
X_train, y_train = create_dataset(train_lstm_scaled, train_lstm_scaled, TIME_STEPS)

X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))

X_train.shape, y_train.shape
```

#### Modeling

All we need to do now is to define the layers of our model and configure it so that we could use it in the prediction phase.

```python
model_lstm = keras.Sequential()
model_lstm.add(
  keras.layers.Bidirectional(
    keras.layers.LSTM(
      units=128,
      input_shape=(X_train.shape[1], X_train.shape[2])
    )
  )
)
model_lstm.add(keras.layers.Dropout(rate=0.2))
model_lstm.add(keras.layers.Dense(units=1))
model_lstm.compile(loss='mean_squared_error', optimizer='adam')
```

Let's fit it now:

```python
model_lstm.fit(
    X_train, y_train,
    epochs=50,
    batch_size=1,
    validation_split=0.1,
    shuffle=False
)
```

#### Predictions

In order to use our model to make predictions, we need first to prepare our testing dataset and scale it.

```python
total= df[["PIB (MDH)"]]

#our testing dataset should include also the values of the previous values that we ll use in the prediction
inputs = total[len(total) - len(test_lstm) - TIME_STEPS:].values
inputs = inputs.reshape(-1,1)

# Now we should scale our data
inputs = sc.transform(inputs)

# Finally, we should prepare the sequences
X_test = []
for i in range(TIME_STEPS, len(inputs)):
    X_test.append(inputs[i-TIME_STEPS:i, 0])
X_test = np.array(X_test)
X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))
```

`X_test` now contains the necessary sequences that we need to make predictions. Let's do that:

```python
# make a predictions 
preds_lstm = model_lstm.predict(X_test)
preds_lstm = sc.inverse_transform(preds_lstm)
preds_lstm = np.reshape(preds_lstm, (preds_lstm.shape[0],))
preds_lstm = pd.DataFrame(preds_lstm,
                         columns=["LSTM PREDS"],
                         index = test_lstm.index[len(test_lstm)-len(preds_lstm):])
# preds_lstm
preds_lstm
```

```python
plt.figure()
plt.plot(train_lstm, label="training")
plt.plot(test_lstm, label = "testing")
plt.plot(preds_lstm, label="predicted (LSTM)")
plt.plot(preds_arima, label="predicted (ARIMA)")
plt.legend(loc = 'upper left')
plt.show()
```

### Stationarity

In order to use the VAR model, we need to check the stationarity of our data since this model requires this kind of data to work.

We already saw in the section of the ARIMA model that the column *PIB (MDH)* is not stationary but needs one integration to make it stationary. Following the same approach, we are going to check the stationarity of the second column *TCR* using the **Advanced Dicky Fuller test** (`sts.adfuller()`)

```python
# Check the stationarity of the second column
adfuller_test = sts.adfuller(multi_df.TCR, autolag= "AIC")
print("ADF stat is: {}".format(adfuller_test[0]))
print("p-value: {}".format(adfuller_test[1]))
```

We can see that the p-value is greater that **0.05** which allows us to conclude that this column is not stationary. So we can transform this column using the `diff()` method and check the stationarity of the transformed data.

```python
#  Checking the stationarity of the transformed data (.diff())
adfuller_test = sts.adfuller(multi_df.TCR.diff()[1:], autolag= "AIC")
print("ADF stat is: {}".format(adfuller_test[0]))
print("p-value: {}".format(adfuller_test[1]))
```

We can see that the p-value is small enough to say that the transformed data is stationary. This transformed data is the one that we ll use to build our model. Let's store this transformde data in a variable `multi_df_d`.

```python
# Storing the transformed data that we ll use to build the model
# this tranformed data contains the difference of each value of the original dataset and the one that precedes it
# That is why we used delta to name the columns

multi_df_d = multi_df.diff().dropna()
multi_df_d.columns = ["delta PIB (MDH)", "delta TCR"]
multi_df_d.head()
```

Let's plot the transformed data:

```python
# creating a figure with 2 plots
fig , ax = plt.subplots(nrows=2)

# Plotting the first transformed column
ax[0].plot("delta PIB (MDH)", data= multi_df_d, label= "delta PIB (MDH)")
ax[0].legend()

# Plotting the second transformed column
ax[1].plot("delta TCR", data= multi_df_d, label= "delta TCR", c="r")
ax[1].legend();
```

Let's scale this two variables and plot them together to explore the impact of each one of them

```python
# scale the variables
sc = MinMaxScaler()
multi_df_d_scaled = pd.DataFrame(sc.fit_transform(multi_df_d), columns= multi_df_d.columns, index = multi_df_d.index)

multi_df_d_scaled.plot();
```

### Parameter analysis

Now, we can split our data to start building our model. we are going to use the same training size that we used in the first analysis (`train_size`)

```python
train_multi_var, test_multi_var = multi_df_d.iloc[:train_size], multi_df_d.iloc[train_size:]
train_multi_var.shape, test_multi_var.shape
```

The parameter that we need to determine for this model is the order. To do so, we are going to fit differennt models using a range of orders and pick the one with the minimal **AIC value**. Lets do that.

```python
# Construct the model using the training dataset
var_model = VAR(train_multi_var)

# fitting models with different orders and store the results in a list
res_aic = []
for p in range(1,10):
    res = var_model.fit(p)
    res_aic.append(res.aic)
```

Now that we stored the AIC values of the results in the `res_aic` list, let's plot them to see the order with the minimal value of the AIC.

```python
# Plotting the Aic values by the order of each model
plt.plot(list(np.arange(1,10)),res_aic)
plt.xlabel("order")
plt.ylabel("AIC");
```

According to the plot, we can say that the model with the order 8 is the one with the least AIC. so we are going t use the VAR(8) model in this section.

### Fitting the model

All we need to do now is fitting the model using the results of our previous analysis (order = 10)

```python
# Fitting the model
res = var_model.fit(8, ic="aic")
res.summary()
```

Let's use this model to make prediction in the next section.

### Predictions

Now that we fitted our model, we can use that to predict the transformed data. we will use the `forecast` method and feed it the lagged values that we need to predict the values of testing dataset.

```python
# store the order of the model
lag_order = res.k_ar

# predict the values of the transformed data using the previous 8 entries necessary to predict the first value
var_preds = res.forecast(train_multi_var.values[-lag_order:], len(test_multi_var))

# storing the predictions in a datatframe
df_var_pred = pd.DataFrame(data= var_preds,
                           index= test_multi_var.index,
                           columns= test_multi_var.columns)
```

```python
# Splitting the data
m_lstm_train, m_lstm_test = multi_df.iloc[:train_size], multi_df.iloc[train_size:]
m_lstm_train.shape, m_lstm_test.shape
```

Now, we are going to scale our data for better performances. We are going to use the `MinMaxScaler()` for this purpose.

```python
#Scaling our data and store the scaled version of it
sc = MinMaxScaler(feature_range = (0, 1))
m_lstm_train_sc = pd.DataFrame(sc.fit_transform(m_lstm_train), 
                               columns= m_lstm_train.columns, 
                               index= m_lstm_train.index)
m_lstm_train_sc.head()
```

We are going to modify `create_dataset()` function that we used in the first lstm model in order to prepare our sequences with precising the time_steps that we want to include in each sequence.

```python
def create_dataset(X, y, time_steps=1):
    Xs, ys = [], []
    for i in range(len(X) - time_steps):
        v = X.iloc[i:(i + time_steps)].values
        Xs.append(v)        
        ys.append(y.iloc[i + time_steps])
    return np.array(Xs), np.array(ys)
```

Now, let's construct our sequences using the scaled data

### GDP analysis

In this section, the `PIB (MDH)` is going to be our target variable:

```python
# Specifying the TIME_STEPS
TIME_STEPS = 2

# We prepare our training set using the previous function
X_train, y_train = create_dataset(m_lstm_train_sc, m_lstm_train_sc["PIB (MDH)"], TIME_STEPS)
X_train.shape, y_train.shape
```

Now that we prepared our sequences, we can begin our model

#### Modeling

Now that we prepared the sequences, let's begin modeling. We are going to use the same structure that we used in the univariate analysis but including the second variable too.

```python
# Constructing the layers of our model
model_lstm_m = keras.Sequential()
model_lstm_m.add(
  keras.layers.Bidirectional(
    keras.layers.LSTM(
      units=128,
      input_shape=(X_train.shape[1], X_train.shape[2])
    )
  )
)

# Adding a dropout layer
model_lstm_m.add(keras.layers.Dropout(rate=0.2))

# Adding the dense layer
model_lstm_m.add(keras.layers.Dense(units=1))

# Compile the model
model_lstm_m.compile(loss='mean_squared_error', optimizer='adam')
```

Let's fit our model now.

```python
model_lstm_m.fit(
    X_train, y_train,
    epochs=30,
    batch_size=1,
    validation_split=0.1,
    shuffle=False
)
```

#### Predictions

Before beginning the prediction phase, we need first to prepare our testing data

```python
# scale the testing data
m_lstm_test_sc = pd.DataFrame(sc.transform(m_lstm_test), 
                               columns= m_lstm_test.columns, 
                               index= m_lstm_test.index)

# Constructing the sequences
X_test, y_test = create_dataset(m_lstm_test_sc, m_lstm_test_sc["PIB (MDH)"], TIME_STEPS)
X_test.shape, y_test.shape
```

Now that we have our model and our testing data prepared, we can use it to make predictions.

```python
# Since our scaler contains two columns, we are going to duplicate the predictions before inversing them
# And keep just the first elements of the results

m_lstm_preds = model_lstm_m.predict(X_test)
m_lstm_preds = np.repeat(m_lstm_preds, m_lstm_train.shape[1], axis= -1)
m_lstm_preds_inv = sc.inverse_transform(m_lstm_preds)[:,0]

# Store predictions in a dataframe
m_lstm_preds_df = pd.DataFrame(m_lstm_preds_inv,
                         columns=["Multivariate LSTM PREDS"],
                         index = m_lstm_test.index[len(m_lstm_test)-len(m_lstm_preds_inv):])
# preds_lstm
m_lstm_preds_df
```
