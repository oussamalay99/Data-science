```python
# import the necessary libraries
from bs4 import BeautifulSoup as bs
import requests
import re
```

```python
# Define a function that will load the content of the web page and assign it to a variable
def get_data(url):
    html = requests.get(url, verify= False).text
    soup= bs(html, "lxml")
    return soup
```

```python
# Adding the docs from their links and label them using the source of the download
def add_doc(links, source):
    for i, link in enumerate(links):
        ext = link.split(".")[-1]
        with open(f"Docs/{source}_{i+1}.{ext}", "wb") as f:
            content = requests.get(link).content
            f.write(content)
    print("Docs are added successfully!")
```

```python
# Storing the URL in a variable and extract the base and a part it as the source
URL= "https://cfetogo.tg/documents/documents-utiles-modeles-20.html"
source = URL.split("//")[-1].split("/")[0].split(".")[0]
base = "https://" + URL.split("//")[-1].split("/")[0]

# Store the content of the page as the source
soup = get_data(URL)
soup
```

```python
# Select the target elements and retrieve the links of the documents from them
target_elts = soup.find_all("a", string = re.compile("Statuts"))
interim_links = [link["href"] for link in target_elts]

# Complete the full URL of the documents and adding them to the folder Docs/
docs = [base + link for link in interim_links]
add_doc(docs, source)
```
